### Image Brightness Adjustment

create a random grayscale image and increase its brightness.

```python
img = np.random.randint(1, 256, (10, 10))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title("Original")
```

```python
img_bright = np.clip(img + 50, 0, 255)
plt.imshow(img_bright, cmap='grey', vmin=0, vmax=255)
plt.title("Brightened")
```

### Manual Statistics

Manually calculate the mean, variance and standard deviation of a given array. Then calculate the same for normalized array.

```python
def calc_mean(arr):
    return arr.sum() / arr.size

def calc_variance(arr):
    avg = calc_mean(arr)
    squared_diff = (arr - avg) ** 2
    return squared_diff.sum() / arr.size

def calc_std(arr):
    return (calc_variance(arr)) ** 0.5 

def print_statistics(arr, deci=3):
    print(f"{calc_mean(arr):.{deci}f}")
    print(f"{calc_variance(arr):.{deci}f}")
    print(f"{calc_std(arr):.{deci}f}")
```

```python
arr1 = np.random.randint(0, 100, (10, 10))
arr2 = np.random.randint(-1, 1, (100, 100))

print_statistics(arr1)
print()
print_statistics(arr2)
```

```python
def normalize(arr):
    """
    min-max normalizaiton
    useful when we need bounded values like pixel intensities
    """
    min = arr.min()
    max = arr.max()
    return (arr - min) / (max - min)
```

```python
arr3 = np.random.randint(50, high=100, size=(100))
norm_arr = normalize(arr3)
print(calc_mean(norm_arr))
```

### Euclidian Distance Calculator

Calculate euclidian distance between 2 points

```python
def calc_dist(p1, p2):
    p1.resize()
    return p1.dot(p2)
```

```python
p1 = np.array([3, 0])
p2 = np.array([7, 0])
print(calc_dist(p1, p2))
```
