# PROGRAMMING PARADIGMS

There are many programming paradigms. Some of them are:
1. Procedural Programming 
    - give computer a set instructions, called procedures to follow
    - eg. C, Pascal, Unix Shells
2. Object-Oriented Programming
    - organize code as collection of objects
    - eg. Java, Smalltalk
3. Declarative Programming
    - tell the computer what you want, not how to do it
    - eg. SQL
4. Functional Programming
    - decomposes the problem into set of functions, which don't change the state of the program
    - eg. Haskell, ML Family (standard ML, OCaml, and other variants)

# FUNCTINOAL PROGRAMMING IN PYTHON

Functional Programming (FP) is a programming paradigm that treats computation as the evaluation of mathematical functions, avoiding changing state and mutable data. It emphasizes *what* to compuute then how to compute it.

The core idea of functional programming is to break the problem into pure functions. Pure functions are those functions which neither change the state of the program nor produce any side effects (like printing something on the terminal [using ```print(...)```]) or pausing the execution of the program with ```sys.sleep()```

The functions here do not depend on external variables, and for a given input, the output  

## Core Philosophy of Functional Programming

In FP, functions are the primary building blocks. The goal is to write programs by composing pure, predictable functions - much like mathematical functions where `f(x)` always returns the same result for the same `x`.

# KEY FEATURES OF FUNCTIONAL PROGRAMMING

## 1. Pure Functions

A function is "pure" if:
- They always return the same output for the same input.
- They don't modify the external state of the program - i.e., functions do not have any *side effects*.

**side effects**: these are effects of a function that change the state of the program. Some examples of side-effects are:
- printing on the terminal using `print()`
- pausing program execution using `sys.sleep()`

````python
# pure function
def add(a, b):
    return a + b

# impure function - modifies external state by printing
def print_sum(a, b):
    print(a+b)
````

## 2. Immutability

Data is never modified after creation. Instead of changing existing data, new data structures are created.

````python
original = [1, 2, 3]
new_list = original + [4]   # original remains unchanged
````

## 3. First-Class and Higher-Order Functions

Functions are treated as **first-class citizens** - they can be:
- assigned to variables
- passed as arguments
- returned from other functions

## 4. Function Composition

Complex functions are built from combining simpler ones, like a pipeline.

## 5. Referential Transparency

An expression is referentially transparent if it can be replaced by its value without changing program behavior. This is a direct result of using pure functions.

````python
def square(x):
    return x*x

# Both are equivalent
result = square(5) + square(7)
result = 25 + 49
````

## 6. Recursion over Loops

FP avoids mutable loop counters and prefers **recursion** for repetition

````python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n * factorial(n-1)
````

# SUMMARIZATION

## Advantages of FP

| Benefit             | Description                                      |
| ------------------- | ------------------------------------------------ |
| **Predictability**  | Pure functions are easy to reason about and test |
| **Concurrency**     | No shared state = no race conditions             |
| **Modularity**      | Small, reusable functions compose well           |
| **Debuggability**   | Isolated functions are easier to trace           |
| **Maintainability** | Less side-effect-related bugs                    |


## Popular Functional Programming Languages

| Language            | Notes                                     |
