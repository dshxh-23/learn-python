# STRINGS

## Strings Basics and Creation

In Python, a string is an **immutable** sequence of **Unicode** characters used to represent text. Strings are objects of the built-in ```str``` class. Python does not have a separate character type; a single character is simply a string of length 1.

Strings can be created using single quotes, double quotes, or triple quotes. Triple quotes are mainly used for multi-line strings.

````python
# Creating strings
s1 = "Hello"
s2 = 'World'
s3 = """This is
a multi-line
string"""

print(s1, s2)
print(s3)
````

## Immutability, Indexing, and Slicing
Strings are immutable, meaning their contents cannot be changed after creation. Any operation that appears to modify a string actually creates a new string.

Strings support indexing (positive and negative) and slicing to extract substrings. Slicing is safe and does not raise index errors.

````python
s = "Programming"

# Indexing
print(s[0])     # First character
print(s[-1])    # Last character

# Slicing
print(s[0:7])   # Program
print(s[:7])    # Program
print(s[7:])    # ming
print(s[::-1])  # Reverse string

# Immutability demonstration
# s[0] = 'X'   # This would raise an error
````

## Common Operations and Iteration

Strings support concatenation, repetition, membership testing, and iteration. Since strings are sequences, they can be looped over character by character.

````python
a = "Hello"
b = "Python"

print(a + " " + b)
print(a * 2)

print("Py" in b)
print("Java" not in b)

for ch in "AI":
    print(ch)
````

## Built-in String Methods (Core Set)

Python provides many built-in methods for working with strings. All string methods return new strings. These methods can be used for case conversion, trimming, searching, replacing, and counting, etc.

The documentation for all the string methods can be found [here](https://docs.python.org/3/library/stdtypes.html#text-sequence-type-str).

````python
s = "  Machine Learning  "

# Case conversion
print(s.lower())
print(s.upper())
print(s.title())

# Trimming whitespace
print(s.strip())

# Searching and counting
print(s.find("Learn"))
print(s.count("n"))

# Replacing
print(s.replace("Machine", "Deep"))
````

## Splitting, Joining, and Validation

Splitting breaks a string into a list based on a delimiter, while joining combines iterable elements into a single string. Validation methods check the nature of string contents.

**NOTE:** Because strings are immutable, repeated concatenation in loops is inefficient. Prefer collecting parts in a list and using ```join()```.

````python
s = "data,science,ml,ai"

parts = s.split(",")
print(parts)

joined = " | ".join(parts)
print(joined)

# Validation methods
print("abc".isalpha())
print("123".isdigit())
print("abc123".isalnum())
print("   ".isspace())
````

## String Formatting and Escape Handling

- Python supports multiple formatting techniques, with f-strings being the most readable and efficient.
- Escape sequences allow special characters like newlines and tabs.
- Raw strings ignore escape processing.

Escape sequences for strings:
- ```\'``` --> '
- ```\'``` --> "
- ```\\``` --> \ 
- ```\t``` --> horizontal tab
- ```\b``` --> backspace
- ```\n``` --> new line

````python
name = "Alice"
age = 21

print(f"My name is {name} and I am {age} years old.")   # using f-string
print("Next year I will be {}.".format(age + 1))        # using formatted string
path = r"C:\Users\Name\Documents"                       # using raw string
print(path)                       

# Escape sequences
print("Line1\nLine2")
print("Tabbed\tText")
````
