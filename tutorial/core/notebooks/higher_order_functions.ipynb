# FIRST-CLASS FUNCITONS

**First-Class Functions:** A programming language is said to have first-class functions if it treats functions as first-class citizens.

**First-Class Citizen:**
- Also called first-class objects
- an entity that supports all operations generally available to other entities
- These operations typically include being passed as an argument, returned from an function, and assigned to a variable

```python
def square(x):
    return x*x

f = square
print(f(5))
```

# HIGHER-ORDER FUNCITONS

**Higher-order Functions:** A funciton is higher order if it does atleast one of the following:
- accept another function as argument
- return a function as a result

The term "higher-order" comes from mathematics, it operates at a *higher level of abstraction*, working on functions rather then raw data. Consider the following higher order function:

```python
def square(n):
    return n*n

def apply_twice(fn, val):
    return fn(fn(val))

print(apply_twice(square, 3))
```

**NOTE:** `apply_twice` does not know or care what `fn` does. `fn` can be anything. Python's `map()`, `filter()` and `sorted()` are all higher order functions

## Accepting functions as arguments

```python
def square(n):
    return n*n

nums = [1, 2, 3, 4, 5]

squares = list(map(square, nums))
print(squares)

evens = list(filter(lambda x: x%2 == 0, nums))
print(evens)

words = ['bananas', 'mango', 'orange']
sorted_words = sorted(words, key=len)   # we pass the built-in funciton len() as an argument to sort strings in order of their lengths
print(sorted_words)

```

## Returning functions as output

```python
def make_multiplier(factor):
    
    def multiplier(n):
        return n * factor

    return multiplier       # returning the inner functions itself, not calling it

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))
print(triple(5))
```

`make_multiplier` is a **function factory**, manufacturing custom functions on demand. You call it once to create a tool, and use the tool repeatedly.

# CLOSURES

**Closures:** A closure is an inner funciton that *remembers* the scope (envoronment) where it was defined in, even when the outer scope has finished execution.

In the above `make_mulitplier` example, notice that once `make_multipler` funciton is called while making `double`, the variable `factor` should be lost. However, the inner function `multiplier` stores the variable `factor` in it's closure, allowing itself to access `factor` whenever required.

A more practical closure is a counter, as follows:

```python
def make_counter():
    count = 0       # this variable lives in closure

    def increment():
        nonlocal count
        count += 1
        return count

    return increment

counter_1 = make_counter()

print(counter_1())  # 1
print(counter_1())  # 2
print(counter_1())  # 3

# A completely separate counter - its own closure, it's own count
counter_2 = make_counter()
print(counter_2())  # 1
```
